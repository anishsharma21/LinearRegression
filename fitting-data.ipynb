```python
from LinearRegression import LinearRegression
from DataGenerator import DataGenerator
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = pd.Series(range(0, 101))
data_generator = DataGenerator(x)
y = data_generator.generate_y()
plt.scatter(x, y)
```

```python
y = pd.Series(y.flatten())
```

```python
lr = LinearRegression()
points = lr.x_y_to_points(x, y)
```

```python
lr.learning_rate = 0.0001
lr.iterations = 1000
lr.crude_fit()
print(lr.b)
print(lr.m)
print(lr.iterations)
```

```python
y_predict = np.array([lr.m * x_val + lr.b for x_val in x]).reshape(-1,1)
plt.scatter(x, y)
plt.plot(x, y_predict)
```

```python
lr.standard_fit()
print(lr.b)
print(lr.m)
print(lr.iterations)
```
